## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python


# Connect to the SQLite database
conn = sqlite3.connect("my_database.sqlite")

# Create a cursor object using the cursor() method
cursor = conn.cursor()

# SQL query to count the number of rows in the chicago_crime_data table
query = "SELECT COUNT(*) FROM chicago_crime_data"

# Execute the query
cursor.execute(query)

# Fetch the result
total_crimes = cursor.fetchone()[0]

# Print the result
print(f"Total number of crimes: {total_crimes}")

# Close the database connection
conn.close()
```

### Problem 2

##### List community areas with per capita income less than 11000.


```python


# Connect to the SQLite database
conn = sqlite3.connect("my_database.sqlite")

# Create a cursor object using the cursor() method
cursor = conn.cursor()

# SQL query to select community areas with per capita income less than 11000
query = """
SELECT community_area_name, per_capita_income 
FROM chicago_socioeconomic_data
WHERE per_capita_income < 11000;
"""

# Execute the query
cursor.execute(query)

# Fetch all the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(f"Community Area: {row[0]}, Per Capita Income: {row[1]}")

# Close the database connection
conn.close()
```

### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)



```python
# Connect to the SQLite database
conn = sqlite3.connect("my_database.sqlite")

# Create a cursor object using the cursor() method
cursor = conn.cursor()

# SQL query to select case numbers for crimes involving minors
# Replace 'CRIME_TYPE' with the actual crime type or description related to minors
query = """
SELECT CASE_NUMBER FROM chicago_crime_data WHERE PRIMARY_TYPE = 'CRIME_TYPE' OR DESCRIPTION LIKE '%keyword%';
"""

# Execute the query
cursor.execute(query)

# Fetch all the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(f"Case Number: {row[0]}")

# Close the database connection
conn.close()
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("my_database.sqlite")

# Create a cursor object using the cursor() method
cursor = conn.cursor()

# SQL query to select case numbers for kidnapping crimes involving a child
query = """
SELECT CASE_NUMBER FROM chicago_crime_data 
WHERE PRIMARY_TYPE = 'KIDNAPPING' AND DESCRIPTION LIKE '%CHILD%';
"""

# Execute the query
cursor.execute(query)

# Fetch all the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(f"Case Number: {row[0]}")

# Close the database connection
conn.close()
```

### Problem 5

##### What kinds of crimes were recorded at schools?


```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("my_database.sqlite")

# Create a cursor object using the cursor() method
cursor = conn.cursor()

# SQL query to select distinct crime types at school locations
query = """
SELECT DISTINCT PRIMARY_TYPE 
FROM chicago_crime_data 
WHERE LOCATION_DESCRIPTION LIKE '%SCHOOL%';
"""

# Execute the query
cursor.execute(query)

# Fetch all the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(f"Crime Type: {row[0]}")

# Close the database connection
conn.close()
```

### Problem 6

##### List the average safety score for each type of school.


```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("my_database.sqlite")

# Create a cursor object using the cursor() method
cursor = conn.cursor()

# SQL query to compute the average safety score for each type of school
query = """
SELECT "Elementary, Middle, or High School", AVG(SAFETY_SCORE) AS Average_Safety_Score
FROM chicago_public_schools
GROUP BY "Elementary, Middle, or High School";
"""

# Execute the query
cursor.execute(query)

# Fetch all the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(f"School Type: {row[0]}, Average Safety Score: {row[1]}")

# Close the database connection
conn.close()
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line 


```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("my_database.sqlite")

# Create a cursor object using the cursor() method
cursor = conn.cursor()

# SQL query to find top 5 community areas with highest % of households below poverty line
query = """
SELECT community_area_name, percent_households_below_poverty 
FROM chicago_socioeconomic_data 
ORDER BY percent_households_below_poverty DESC 
LIMIT 5;
"""

# Execute the query
cursor.execute(query)

# Fetch all the results
results = cursor.fetchall()

# Print the results
for row in results:
    print(f"Community Area: {row[0]}, % Households Below Poverty Line: {row[1]}")

# Close the database connection
conn.close()
```

### Problem 8

##### Which community area is most crime prone?


```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("my_database.sqlite")

# Create a cursor object using the cursor() method
cursor = conn.cursor()

# SQL query to find the community area with the highest number of crimes
query = """
SELECT COMMUNITY_AREA_NUMBER, COUNT(*) as Crime_Count
FROM chicago_crime_data
GROUP BY COMMUNITY_AREA_NUMBER
ORDER BY Crime_Count DESC
LIMIT 1;
"""

# Execute the query
cursor.execute(query)

# Fetch the result
result = cursor.fetchone()

# Print the result
print(f"Most Crime Prone Community Area: {result[0]}, Number of Crimes: {result[1]}")

# Close the database connection
conn.close()
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("my_database.sqlite")

# Create a cursor object using the cursor() method
cursor = conn.cursor()

# SQL query to find the name of the community area with the highest hardship index
query = """
SELECT Community_Area_Name
FROM chicago_socioeconomic_data
WHERE Hardship_Index = (
    SELECT MAX(Hardship_Index)
    FROM chicago_socioeconomic_data
);
"""

# Execute the query
cursor.execute(query)

# Fetch the result
result = cursor.fetchone()

# Print the result
print(f"Community Area with the Highest Hardship Index: {result[0]}")

# Close the database connection
conn.close()
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("my_database.sqlite")

# Create a cursor object using the cursor() method
cursor = conn.cursor()

# SQL query to find the community area name with the most number of crimes
query = """
SELECT Community_Area_Name
FROM chicago_socioeconomic_data
WHERE Community_Area_Number = (
    SELECT Community_Area_Number
    FROM chicago_socioeconomic_data
    GROUP BY Community_Area_Number
    ORDER BY COUNT(*) DESC
    LIMIT 1
);

"""

# Execute the query
cursor.execute(query)

# Fetch the result
result = cursor.fetchone()

# Print the result
print(f"Community Area with the Most Crimes: {result[0]}")

# Close the database connection
conn.close()
```
